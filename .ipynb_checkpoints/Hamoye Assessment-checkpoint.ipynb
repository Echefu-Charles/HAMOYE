```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://github.com/WalePhenomenon/climate_change/blob/master/fuel_ferc1.csv?raw=true'
df = pd.read_csv(url, error_bad_lines=False)

df2 = pd.read_csv('https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv', error_bad_lines=False)
```

```python
print(df.shape)
print(df2.shape)
```

```python
# Features description
'''
'Record_id' : record id

'Utility_id_ferc1': Utility id assigned by the FERC

'Report_year': year of report

'Plant_name_ferc1': the name of the plant

'Fuel_type_code_pudl': the type of fuel

'Fuel_unit': the unit of fuel

'Fuel_qty_burned': the quantity of fuel burned

'Fuel_mmbtu_per_unit': the measure of energy per unit

'fuel_cost_per_unit_burned': the fuel cost per unit burned

'Fuel_cost_per_unit_delivered': the cost of fuel delivered per unit

'fuel_cost_per_mmbtu': the cost of fuel per mmbtu
'''
```

```python
sns.countplot(x = 'fuel_type_code_pudl', data = df)
```

```python
plt.figure(figsize=(30, 10))
sns.countplot(x = 'report_year', hue='fuel_type_code_pudl', data = df)
plt.xlabel('Report Years')
plt.title('Plot of Fuel type usage across years')
plt.show()
```

```python
fuel_per_year = df.groupby('report_year').agg({'fuel_qty_burned':'sum',
       'fuel_mmbtu_per_unit':'sum', 'fuel_cost_per_unit_burned':'sum',
       'fuel_cost_per_unit_delivered':'sum', 'fuel_cost_per_mmbtu':'sum'})
```

```python
fuel_type = df.groupby('fuel_type_code_pudl').agg({'fuel_qty_burned':'sum',
       'fuel_mmbtu_per_unit':'sum', 'fuel_cost_per_unit_burned':'sum',
       'fuel_cost_per_unit_delivered':'sum', 'fuel_cost_per_mmbtu':'sum'})
```

```python
fuel_type
```

```python
fuel_per_year
```

```python
plt.figure(figsize=(20, 10))
plt.plot(fuel_per_year.index, 'fuel_qty_burned', data = fuel_per_year, color='green', marker='o', linestyle='dashed',
         linewidth=2, markersize=12)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(fuel_per_year.index, 'fuel_mmbtu_per_unit', data = fuel_per_year, color='blue', marker='o', linestyle='dashed',
         linewidth=2, markersize=12)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(fuel_per_year.index, 'fuel_cost_per_unit_burned', data = fuel_per_year, color='black', marker='o', linestyle='dashed',
         linewidth=2, markersize=12)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(fuel_per_year.index, 'fuel_cost_per_unit_delivered', data = fuel_per_year, color='red', marker='o', linestyle='dashed',
         linewidth=2, markersize=12)
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(fuel_per_year.index, 'fuel_qty_burned', data = fuel_per_year, color='green', marker='o', linestyle='dashed',
         linewidth=2, markersize=12)
ax.plot(fuel_per_year.index, 'fuel_mmbtu_per_unit', data = fuel_per_year, color='blue', marker='o', linestyle='dashed',
         linewidth=2, markersize=12)
ax.plot(fuel_per_year.index, 'fuel_cost_per_unit_burned', data = fuel_per_year, color='black', marker='o', linestyle='dashed',
         linewidth=2, markersize=12)
ax.plot(fuel_per_year.index, 'fuel_cost_per_unit_delivered', data = fuel_per_year, color='red', marker='o', linestyle='dashed',
         linewidth=2, markersize=12)
plt.grid(True)
plt.show()
```
